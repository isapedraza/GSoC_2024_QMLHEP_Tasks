```python
import numpy as np

# Carga el archivo npz con allow_pickle=True para permitir la carga de objetos
with np.load('/content/drive/MyDrive/WORK_2024/CIIEC/Proyectos/ComputacionCuantica/VeranoCMS/QIS_EXAM_200Events.npz', allow_pickle=True) as data:
    # Lista todas las claves del archivo
    keys = data.files
    print("Claves en el archivo npz:", keys)

    # Acceder a cada objeto por su clave y manejar según su tipo
    for key in keys:
        obj = data[key]
        print(f"Clave: {key}, Tipo de dato: {type(obj)}")

        # Intenta manejar el objeto según su tipo
        if hasattr(obj, 'shape'):
            # Si el objeto es un array de NumPy con una forma definida
            print(f"Forma: {obj.shape}")
            if obj.size > 0:  # Asegura que el objeto no esté vacío
                # Ajusta la impresión si es un array
                num_elements = min(obj.size, 10)  # Limita a 10 elementos
                print(f"Primeros {num_elements} elementos de '{key}':", obj.flat[:num_elements])
        else:
            # Si el objeto no es un array de NumPy o no tiene una forma definida
            print(f"Contenido de '{key}': {obj}")

```
